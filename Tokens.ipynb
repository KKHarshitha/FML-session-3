Tokenization using NLTK:

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

#word tokenize

text = '''Sentiment analysis is the practice of using algorithms to classify various samples of related text into overall positive and negative categories.'''
tokens = word_tokenize(text)
print(tokens)
print("No.Of tokens",len(tokens))
```

```python

#sent 
from nltk.tokenize import sent_tokenize
sentances = sent_tokenize(text)
print(sentances)
print("No.Of sentance",len(sentances))
```

Tokenization using Pythons split() function:

```python
#word tokenization
tokens = text.split()
print(tokens)
print("No.Of tokens",len(tokens))
```

```python
#sent
sentances = text.split(".")
print(sentances)
print("No.Of sentances",len(sentances))
```

Tokenization using Regular Expressions (RegEx):

```python
import re
#word tokenization
tokens = re.findall("[\w']+",text)
print(tokens)
print("no.of tokens",len(tokens))
```

```python
#sent 
sentences = re.compile('[.?!] ').split(text)
print(sentences)
print("No.of sentences : ", len(sentences))
```
